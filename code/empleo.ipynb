```python
salarios = sal[['RETRINOIN','ESTU','SEXO','CNACE']]
```

```python
salarios.sort_values(by='RETRINOIN', inplace=True)
salarios
```

```python
salarios = salarios[salarios['RETRINOIN']!=0]
```

```python
ccc = salarios.groupby('CNACE').agg({'RETRINOIN':'mean'}).reset_index()
```

```python
ddd = salarios.groupby('CNACE').agg({'RETRINOIN':'median'}).reset_index()
```

```python
ccc.rename(columns={'RETRINOIN':'Media'}, inplace=True)
```

```python
ddd.rename(columns={'RETRINOIN':'Mediana'}, inplace=True)
```

```python
ddd = ddd.drop(columns='CNACE')
```

```python
ccc = pd.concat((ccc, ddd),axis=1)
ccc
```

```python
rows_C = ccc[ccc['CNACE'].str.match('C[1-8]')]
mean_C = rows_C['Media'].mean()
median_C = rows_C['Mediana'].median()

ccc.loc[len(ccc)] = ['C0', mean_C, median_C]
```

```python
rows_G = ccc[ccc['CNACE'].str.match('G[1-2]')]
mean_G = rows_G['Media'].mean()
median_G = rows_G['Mediana'].median()

ccc.loc[len(ccc)] = ['G0', mean_G, median_G]
```

```python
rows_H = ccc[ccc['CNACE'].str.match('H[1-2]')]
mean_H = rows_H['Media'].mean()
median_H = rows_H['Mediana'].median()

ccc.loc[len(ccc)] = ['H0', mean_H, median_H]
```

```python
intervalo = [x for x in range(0, 4226000, 1000)]

def encontrar_cercano(valor):
    return min(intervalo, key=lambda x:abs(x-valor))

salarios['rango_cercano'] = salarios['RETRINOIN'].apply(lambda x: encontrar_cercano(x))

salarios
```

```python
intervalo = [x for x in range(0, 4226000, 500)]

def encontrar_cercano(valor):
    return min(intervalo, key=lambda x:abs(x-valor))

salarios['rango_cercano_500'] = salarios['RETRINOIN'].apply(lambda x: encontrar_cercano(x))

salarios
```
